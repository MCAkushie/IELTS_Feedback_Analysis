```python
# importing NLP libraries for sentiment labelling and preprocessing
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#Initializing sentiment Analyzer for sentiment labelling
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return sentiment


for column in df.columns:

    print(column)
    if (column != 'Significance Rating'):
        sentiment_column = column + ' sentiment'
        df[sentiment_column] = df[column].apply(get_sentiment)
```

```python
df.head()
```

```python
#Assigning parameters for sentiment labelling
def determine_value(row):
    if row['Feedback on Necessity sentiment'] == 1 and row['Significance Rating'] > 3:
        return 'positive'
    elif row['Feedback on Necessity sentiment'] == 0 and row['Significance Rating'] < 3:
        return 'negative'
    else:
        return 'neutral'
    
def determine_polarity(row):
    if row['Feedback on Necessity sentiment'] == 1 and row['Significance Rating'] > 3:
        return 1
    elif row['Feedback on Necessity sentiment'] == 0 and row['Significance Rating'] < 3:
        return -1
    else:
        return 0
    
df['Overall sentiment'] = df.apply(determine_value, axis=1)
df['Overall sentiment value'] = df.apply(determine_polarity, axis=1)
```

```python
#Importing relevant libraries
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt

# Define the stop words
stopwords = set(STOPWORDS)

# Function to preprocess text
def preprocess_text(text):
    # Remove special characters, numbers, and punctuations
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = re.sub(r'\d+', '', text)
    
    # Tokenize the text
    words = text.split()
    
    # Remove stop words
    words = [word.lower() for word in words if word.lower() not in stopwords]
    
    return ' '.join(words)

# Apply the preprocessing function
df['feedback_lowercase'] = df['feedback_lowercase'].apply(preprocess_text)

#Joining all feedback text into a single string
textt = " ".join(text for text in df.feedback_lowercase)

#Generating word cloud
wordcloud = WordCloud(stopwords=stopwords).generate(textt)

#%matplotlib inline
#Displaying the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Model Evaluation
def confusion_metrics(conf_matrix):

    TNg = conf_matrix     [0][0]  # True Negative
    FNt_neg = conf_matrix [0][1]  # False Neutral (actual negative predicted as neutral)
    FP = conf_matrix      [0][2]  # False Positive
    FNg = conf_matrix     [1][0]  # False Negative (actual neutral predicted as negative)
    TNt = conf_matrix     [1][1]  # True Neutral
    FP_neu = conf_matrix  [1][2]  # False Positive (actual neutral predicted as positive)
    FN = conf_matrix      [2][0]  # False Negative (actual positive predicted as negative)
    FNt_pos = conf_matrix [2][1]  # False Neutral (actual positive predicted as neutral)
    TP = conf_matrix      [2][2]  # True Positive

    print('True Positives (TP):', TP)
    print('True Negatives (TNg):', TNg)
    print('False Positives (FP):', FP)
    print('False Negatives (FNg):', FNg)
    print('True Neutral (TNt):', TNt)
    print('False Neutral (FNt):', FNt_neg + FNt_pos)

    # Calculate accuracy
    conf_accuracy = float(TP + TNg + TNt) / float(TP + TNg + FP + FNg + TNt + FNt_neg + FNt_pos + FN + FP_neu)

    # Calculate misclassification
    conf_misclassification = 1 - conf_accuracy

    # Calculate the sensitivity (Recall) for Positive
    conf_sensitivity_pos = TP / float(TP + FN)

    # Calculate the sensitivity (Recall) for Negative
    conf_sensitivity_neg = TNg / float(TNg + FP + FNt_neg)

    # Calculate the sensitivity (Recall) for Neutral
    conf_sensitivity_neu = TNt / float(TNt + FNg + FP_neu)

    # Calculate the specificity for Negative
    conf_specificity_neg = TNg / float(TNg + FP)

    # Calculate precision for Positive
    conf_precision_pos = TP / float(TP + FP_neu)

    # Calculate precision for Negative
    conf_precision_neg = TNg / float(TNg + FN + FNt_neg)

    # Calculate precision for Neutral
    conf_precision_neu = TNt / float(TNt + FNg)

    # Calculate f_1 score for Positive
    conf_f1_pos = 2 * ((conf_precision_pos * conf_sensitivity_pos) / (conf_precision_pos + conf_sensitivity_pos))

    # Calculate f_1 score for Negative
    conf_f1_neg = 2 * ((conf_precision_neg * conf_sensitivity_neg) / (conf_precision_neg + conf_sensitivity_neg))

    # Calculate f_1 score for Neutral
    conf_f1_neu = 2 * ((conf_precision_neu * conf_sensitivity_neu) / (conf_precision_neu + conf_sensitivity_neu))

    print('-' * 50)
    print(f'Accuracy: {round(conf_accuracy, 2)}')
    print(f'Mis-Classification: {round(conf_misclassification, 2)}')
    print(f'Sensitivity (Positive): {round(conf_sensitivity_pos, 2)}')
    print(f'Sensitivity (Negative): {round(conf_sensitivity_neg, 2)}')
    print(f'Sensitivity (Neutral): {round(conf_sensitivity_neu, 2)}')
    print(f'Specificity (Negative): {round(conf_specificity_neg, 2)}')
    print(f'Precision (Positive): {round(conf_precision_pos, 2)}')
    print(f'Precision (Negative): {round(conf_precision_neg, 2)}')
    print(f'Precision (Neutral): {round(conf_precision_neu, 2)}')
    print(f'f_1 Score (Positive): {round(conf_f1_pos, 2)}')
    print(f'f_1 Score (Negative): {round(conf_f1_neg, 2)}')
    print(f'f_1 Score (Neutral): {round(conf_f1_neu, 2)}')
```
